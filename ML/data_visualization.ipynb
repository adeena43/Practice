```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")
df.head()

#to see distribution of a numeric feature
sns.histplot(df['total_bill'], bins = 20, kde = True)
plt.title("Distribution of total bill")
plt.show()
```

```python
#Box plot used to detect outliers and compare numeric vs categorical data
sns.boxplot(x = "day", y = "total_bill", data = df)
plt.title("Total bill per day")
plt.show()
```

```python
#violi plot: Like boxplot but also shows data density.
sns.violinplot(x="day", y="total_bill", data=df)
plt.title("Violin Plot of Total Bill per Day")
plt.show()
```

```python
#Count plot: used for counting categorical values
sns.countplot(x = "day", data  = df)
plt.title("Count of Records per Day")
plt.show()
```

```python
#Bar plot
#Used for comparing averages or aggregates of numeric data grouped by categories.
sns.barplot(x="day", y="tip", data=df)
plt.title("Average Tip per Day")
plt.show()
```

```python
#Scatter plot: Used to see relationships between two numeric variables.
sns.scatterplot(x = 'total_bill', y = "tip", data = df)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
#Pair plot: Plots pairwise relationships between numeric columns.
sns.pairplot(data = df)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# line plot:Used to show trends over a continuous variable, usually time (like dates, months, or sequence/order).
sns.lineplot(x = "size", y = "total_bill", data = df)
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=df)
plt.title("Regression Line for Linearity Check")
plt.show()
```
